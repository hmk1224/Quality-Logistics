```python
# import necessary libraries
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import OrdinalEncoder

```

```python
# load the dataset
data = pd.read_csv("logisticQuality.csv")
data.head()
```

```python
obj_df = data.select_dtypes(include=['object']).copy()
ord_enc = OrdinalEncoder()
obj_df["make_code"] = ord_enc.fit_transform(obj_df[["State wise scores of individual parameters"]])
obj_df[["make_code", "State wise scores of individual parameters"]].head(11)

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["State wise scores of individual parameters"] = le.fit_transform(data["State wise scores of individual parameters"])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

# read in the dataset
df = pd.read_csv("logisticQuality.csv")

# split the dataset into input (X) and output (y) variables
X = df.drop(["Final Score","State wise scores of individual parameters"], axis=1)
y = df["Final Score"]

# scale the input variables
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# define the neural network
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

# compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# fit the model to the training data
model.fit(X_train, y_train, epochs=50, batch_size=32)

# evaluate the model on the test data
test_loss = model.evaluate(X_test, y_test)
print("Test Loss:", test_loss)
```

```python
from sklearn.metrics import mean_squared_error

# predict the target variable
y_pred = model.predict(X_test)

# calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# calculate the root mean squared error
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
